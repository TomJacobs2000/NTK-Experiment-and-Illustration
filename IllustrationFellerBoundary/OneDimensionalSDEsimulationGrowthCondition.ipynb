```python
from tqdm import tqdm
import numpy as np
import  matplotlib.pyplot as plt
np.random.seed(13)
```

```python
def f(r):
    'Function evaluation of f on r'
    
    evaluation = r**2 + r**4
    
    return evaluation
```

```python
def gradf(r):
    
    evaluation = 2*r + 4*r**3
    
    return evaluation
```

```python
def gradf2(r):
    
    evaluation =2 + 12*r**2
    
    return evaluation
```

```python
def Sim(n,gamma, eta, T, N, r_init, runs):
    
    h = T/N # stepsize
    
    print(h < eta * f(r_init)) # check if the step is smaller than the order of the noise
    
    LossPath  = []
    
    for j in tqdm(range(runs)):
    
        r_norm_init = np.linalg.norm(r_init)

        # Path1 = [] # store norm of x
        Path2 = []

        # Loss1 = []
        Loss2 = []

        # r1 = r_init
        r2 = r_init
    
        # Path1.append(r1)
        Path2.append(r2)

        for i in range(N):
            
            #b1 = -gradf(r1) + (n-1)/r1 * eta * f(r1)/2
            #s1 = np.sqrt(2*eta*f(r1))

            b2 = -gradf(r2) + (n-1)/r2 * eta * f(r2)
            b2prime = -gradf2(r2) + (n-1)*eta *(-1/r2**2 *f(r2) + gradf(r2)/r2)
            s2 = np.sqrt(2*eta*f(r2))

            dW = np.random.normal(0, 1, 1)
            
            # r1 = r1 + b1*h + np.sqrt(h)*s1*dW # EM
            r2 = r2 + b2*h + np.sqrt(h)*s2*dW + b2*b2prime*(dW**2*h - h)/2 # Milstein

            #Path1.append(r1)
            Path2.append(r2)

            #Loss1.append(f(r1))
            Loss2.append(f(r2))

            if(f(r2) <= 0):
                break
                
        LossPath.append([Loss2])  
        
    return(LossPath)
```

```python
alpha = 0.7
n = 1000# co- dimension of Loss manifold
eta = 1/((n)/2-1)*alpha # Learning rate
gamma = 4 # Growth condition
print(-2-2/eta < -n)

T = 1 # time
N = 100000# steps
r_init = 0.1 # initial position
runs = 10

LossPath = Sim(n,gamma, eta, T, N, r_init, runs)

for i in range(runs):
    plt.plot(LossPath[i][0])
#plt.yscale('log')   
plt.xlabel('k-th itteration')
plt.ylabel('Loss')
plt.ylim(0,0.02)
plt.title(f'$alpha = {alpha}$')
plt.savefig(f'1dSim{alpha}GC.png')
```
